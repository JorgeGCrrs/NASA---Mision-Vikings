# CLUSTERING

### Importamos las librer√≠as

```python
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA as sklearnPCA
```

### Importamos el dataset

```python
filename = "nasa/event/event_wind_summary/event_wind_summary.csv"
fd = pd.read_csv(filename);
```

### Hacemos limpieza de datos y seleccionamos las columnas

```python
fd.columns
```

```python
fd_sel  =  fd[['SEISMIC_TIME_SOLS', 'DATA_ACQUISITION_HOUR','PRESSURE','WINDSPEED','AIR_TEMPERATURE','MEAN_X_AXIS_CROSSINGS','RMS_X_AXIS_X100',
       'RMS_Y_AXIS_X100', 'RMS_Z_AXIS_X100']]
fd_less =  fd_sel.dropna(how='any');
```

### Normalizamos el dataframe para que no exista ninguna columna predominante

```python
df_norm = ( fd_less - fd_less.min()) / ( fd_less.max() - fd_less.min() )
df_norm.mean()
```

## CLUSTERING K-MEANS

```python
usados = ['WINDSPEED','PRESSURE','AIR_TEMPERATURE','RMS_X_AXIS_X100',
       'RMS_Y_AXIS_X100', 'RMS_Z_AXIS_X100' ]
#usados = ['WINDSPEED','AIR_TEMPERATURE','PRESSURE']
```

```python
km = KMeans(n_clusters = 4).fit(df_norm[usados])
lbl = pd.Series(km.labels_)
```

```python
df_norm["TAG_KM"] = lbl; 
```

```python
df_norm.groupby("TAG_KM").mean()
```

```python
df_norm['TAG_KM'].hist()
```

```python
fd_less["TAG_KM"] = lbl;
fd_less.groupby("TAG_KM").max()
```

## Estudio de los grupos

#### GRUPO 0

```python

X = fd_less[fd_less.TAG_KM == 0];

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

#y = np.choose(labels, [1, 2, 0]).astype(np.float)
ax.scatter(X.SEISMIC_TIME_SOLS, X.DATA_ACQUISITION_HOUR,X.RMS_X_AXIS_X100, edgecolor='k')

#ax.scatter(X[:,0],X[:,1],X[:,2], edgecolor='y')
plt.title("Todos los valores maximos del sismografo")
ax.set_xlabel("DATA_ACQUISITION_SOL")
ax.set_ylabel("DATA_ACQUISITION_HOUR")
ax.set_zlabel('MAXIMO VALOR')


plt.show()
```
